```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"RidershipbyRouteTableDownload.csv")
df = df[df['Service Category'] == 'Muni Metro']
df = df[df['Service Day of the Week'] == 'Weekday']
df['Average Daily Boardings'] = df['Average Daily Boardings'].str.replace(',', '').astype(float)
df['Average Daily Boardings'] = pd.to_numeric(df['Average Daily Boardings'])
df19 = df[df['Month'].astype(str).str.contains('19')]
df25 = df[df['Month'].astype(str).str.contains('25')]
df19.to_csv('2019MuniMetroRidershipWeekdays.csv')
df25.to_csv('2025MuniMetroRidershipWeekdays.csv')
```

```python
print(df.columns)
print(df19['Route'].unique())
print(df['Average Daily Boardings'].dtype)
```

```python
#Calculate and Plot Average Ridership by Route in 2025
Routes = ['J Church', 'K Ingleside', 'KT Ingleside/Third', 'L Taraval', 
          'M Ocean View', 'N Judah', 'T Third']

average_data = {}

for route in Routes:
    route_data = df25[df25['Route'] == route]
    avg_daily = route_data['Average Daily Boardings'].mean(skipna=True)
    avg_data[route] = avg_daily
    print(f"Average Ridership for {route} in 2025: {avg_daily:.2f}")

avg_series = pd.Series(avg_data)
avg_series = avg_series.sort_values()

colors = ['orange', '#FA93E8', '#87CCFF','#8A00FF','#53AB50','#CF2D2D','#1E34B3']

plt.figure(figsize=(10, 6))
avg_series.plot(kind = 'bar', color=colors)
plt.title('Average Daily Ridership by Route (2025)')
plt.ylabel('Average Daily Boardings')
plt.xlabel('Routes')
plt.xticks(rotation=27)
plt.tight_layout()
plt.savefig('Average Daily Ridership by Route (2025).png')
plt.show()
```
